### Probability mass function of a discrete random variable

```python
p_1 = 0.65
p_2 = 0.29
p_3 = 0.05
p_4 = 0.007
p_5 = 1 - (p_1 + p_2 + p_3 + p_4)
```

```python
pmf = {
    0: p_1,
    1: p_2,
    2: p_3,
    3: p_4,
    4: p_5,
}
```

```python
values = [p_1, p_2, p_3, p_4, p_5]
names = [fr'$x_{i+1}={i}$' for i in range(5)]

fig, axs = plt.subplots(1, 2, figsize=(6, 3), sharey=True)


axs[0].scatter(names, values)
axs[1].step(names, values)
fig.suptitle("Cumulative distibution function of a disrete random variable")
plt.show()
```

## Функция на плътността на вероятността на гаусово разпределени

```python
x = np.linspace(-3, 3, 1000)
x_mean = 0
x_std = 1
pdf = 1 / (x_std * np.sqrt(2 * np.pi)) * np.exp(-0.5 * ((x - x_mean) / x_std) ** 2)
```

```python
plt.plot(x, pdf)
plt.show()
```
