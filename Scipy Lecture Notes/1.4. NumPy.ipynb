#### Worked Example: diffusion using a random walk algorithm

```python
lanes = 1000  # number of walkers
t_max = 250  # time during which we follow the walker
```

```python
steps = np.random.choice([-1, 1], (lanes, t_max))
steps.shape  # Verification: shape is correct
```

```python
np.unique(steps)  # Verification: all steps are -1 or 1
```

```python
positions = steps.cumsum(axis=1)
positions
```

```python
sq_distance = positions ** 2
sq_distance[:5]
```

```python
mean_sq_distance = sq_distance.mean(axis=0)
mean_sq_distance[:5]
```

```python
plt.figure()
t = np.arange(steps.shape[1])
distance = np.sqrt(mean_sq_distance)
plt.plot(t, distance, 'go', t, np.sqrt(t), 'r--')
plt.xlabel(r"$t$") 
plt.ylabel(r"$\sqrt{\langle (\delta x)^2 \rangle}$") 
plt.show()
```

```python
plt.plot(t, mean_sq_distance, 'go', t, t, 'r--')
plt.show()
```
