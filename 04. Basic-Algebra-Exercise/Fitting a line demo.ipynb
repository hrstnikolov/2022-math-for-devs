```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
```

```python
RANDOM_STATE = 42
```

# Fitting a line demo

```python
X , _ = make_blobs(centers=[[1, 3], [2, 1.5], [3, 1.3]], cluster_std=[0.5, 0.35, 0.2], random_state=RANDOM_STATE)
x, y = X[:, 0], X[:, 1]
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
coeffs = np.polyfit(x, y, 2)
```

```python
f = np.poly1d(coeffs)
```

```python
fitted_x = np.linspace(0, 4, 100)
fitted_y = f(fitted_x)
```

```python
plt.scatter(x, y)
plt.plot(fitted_x, fitted_y, c='r')
plt.annotate('fitted 2nd degree polynomial line', xy=[2, f(2)], xytext=[2,3], arrowprops={'arrowstyle': '->'} )

plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
def fit_line(degree):
    coeffs = np.polyfit(x, y, degree)

    f = np.poly1d(coeffs)

    fitted_x = np.linspace(0, 4, 100)
    fitted_y = f(fitted_x)

    plt.scatter(x, y)
    plt.plot(fitted_x, fitted_y, c='r')
    plt.annotate(f'fitted {degree} degree polynomial line', xy=[2, f(2)], xytext=[2,3], arrowprops={'arrowstyle': '->'} )

    plt.xlabel('x')
    plt.ylabel('y')
    
fit_line(3)
```

```python
fit_line(4)
```

```python
fig, axs = plt.subplots(4, 2, figsize=(8, 20))

for i, ax in enumerate(axs.flatten()):
    degree = 1 + i
    plt.sca(ax)
    fit_line(degree)
```
