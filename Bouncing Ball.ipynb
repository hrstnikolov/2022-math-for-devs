```python
import numpy as np
import matplotlib.pyplot as plt
```

# Bouncing Ball

## Projectile Motion

```python
# Constants
g = 9.8  # m / (s ** 2)
```

```python
# Initial condtions
x0 = 0
y0 = 0
v0 = 40  # initial velocity, m / s
theta = np.pi / 4  # angle, rad
```

```python
# Equations for velocity
t = np.linspace(0, 5)
vx = np.cos(theta) * v0
vy = np.sin(theta) * v0 - g * t
```

```python
plt.plot(t, vy, label="Vy")
plt.show()
```

```python
# Equations for displacement
rx = x0 + v0 * np.cos(theta) * t
ry = y0 + v0 * np.sin(theta) * t - 0.5 * g * t ** 2
```

```python
plt.plot(rx, ry, label="trajectory")
plt.xlabel("Distance, m")
plt.ylabel("Height, m")
plt.legend()
plt.show()
```
