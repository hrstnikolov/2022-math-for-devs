# Quantiles

## Example

```python
s = np.array([3, 6, 7, 8, 8, 10, 13, 15, 16, 20])
```

```python
quartiles = np.quantile(s, [0.25, 0.5, 0.75])
quartiles
```

```python
plt.plot(np.ones_like(s), s, 'bo')
for q in quartiles:
    plt.axhline(q, color='red')
    
plt.show()
```
